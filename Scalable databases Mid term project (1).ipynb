# SQL queries 

```python
import sqlite3
import pandas as pd

# Connect to SQLite database
conn = sqlite3.connect('Demographic_snapshot_school.db')

# Load datasets into pandas dataframes
dataset1 = pd.read_csv('./2013_-_2018_Demographic_Snapshot_School_20231116.csv')
dataset2 = pd.read_csv('./2019-20_Demographic_Snapshot_-_School_20231116.csv')
dataset3 = pd.read_csv('./2020-2021_Demographic_Snapshot_School_20231116.csv')

# Create tables in the SQLite database
dataset1.to_sql('dataset1_table', conn, index=False, if_exists='replace')
dataset2.to_sql('dataset2_table', conn, index=False, if_exists='replace')
dataset3.to_sql('dataset3_table', conn, index=False, if_exists='replace')

# Query 1: Select all columns from the first dataset with updated column names
query1 = "SELECT SchoolName, TotalEnrollment, Poverty FROM dataset1_table;"

# Execute queries
result1 = pd.read_sql_query(query1, conn)
# Print results
print("Query 1:")
print(result1)
```

```python
import sqlite3
import pandas as pd

# Connect to SQLite database
conn = sqlite3.connect('Demographic_snapshot_school.db')

# Query 2: Filter data from the second dataset for schools with enrollments greater than 1000 with updated column names
query2 = "SELECT SchoolName, TotalEnrollment, Poverty FROM dataset2_table WHERE TotalEnrollment > 1000;"

result2 = pd.read_sql_query(query2, conn)
print("\nQuery 2:")
print(result2)
```

```python
import sqlite3
import pandas as pd

# Connect to SQLite database
conn = sqlite3.connect('Demographic_snapshot_school.db')


# Query 3: Aggregate total enrollments and poverty count for each school across all datasets 
query3 = """
    SELECT
        SchoolName,
        SUM(TotalEnrollment) AS TotalEnrollment,
        SUM(Poverty) AS Poverty
    FROM
        (
            SELECT SchoolName, TotalEnrollment, Poverty FROM dataset1_table
            UNION ALL
            SELECT SchoolName, TotalEnrollment, Poverty FROM dataset2_table
            UNION ALL
            SELECT SchoolName, TotalEnrollment, Poverty FROM dataset3_table
        ) AS combined_data
    GROUP BY SchoolName;
"""
result3 = pd.read_sql_query(query3, conn)
print("\nQuery 3:")
print(result3)
```

```python
import sqlite3
import pandas as pd

# Connect to SQLite database
conn = sqlite3.connect('Demographic_snapshot_school.db')


# Query 4: Join datasets on school_name and select schools with high poverty from the third dataset with updated column names
query4 = """
    SELECT
        d3.SchoolName,
        d3.TotalEnrollment AS TotalEnrollment,
        d3.Poverty AS Poverty
    FROM
        dataset3_table AS d3
    JOIN
        dataset1_table AS d1 ON d1.SchoolName = d3.SchoolName
    WHERE
        d3.Poverty > 50;
"""
result4 = pd.read_sql_query(query4, conn)
print("\nQuery 4:")
print(result4)

# Close database connection
conn.close()
```

```python
import pandas as pd

# Connect to SQLite database
conn = sqlite3.connect('Demographic_snapshot_school.db')

# Query: Find schools with the highest total enrollment across all datasets
query5 = """
    SELECT
        SchoolName,
        MAX(TotalEnrollment) AS MaxTotalEnrollment
    FROM
        (
            SELECT SchoolName, TotalEnrollment FROM dataset1_table
            UNION ALL
            SELECT SchoolName, TotalEnrollment FROM dataset2_table
            UNION ALL
            SELECT SchoolName, TotalEnrollment FROM dataset3_table
        ) AS combined_data
    GROUP BY SchoolName
    ORDER BY MaxTotalEnrollment DESC
    LIMIT 5;  -- Limiting to the top 5 schools for illustration
"""

# Execute the query
result5 = pd.read_sql_query(query5, conn)

# Print the result
print("\nQuery 5:")
print(result5)

# Close database connection
conn.close()
```

Relevant columns ('Poverty' and 'Year') are extracted from three different datasets (df1, df2, df3).

Peaks or shifts in the distribution may indicate changes in the poverty levels over time or differences between datasets.

```python
import pandas as pd
import seaborn as sns


# Extract relevant columns
relevant_cols_poverty = ['Poverty', 'Year']

# Concatenate the datasets
combined_data_poverty = pd.concat([df1[relevant_cols_poverty], df2[relevant_cols_poverty], df3[relevant_cols_poverty]], keys=['df1', 'df2', 'df3'])

# Define a function to check if a value is numeric
def is_numeric(value):
    try:
        float_value = float(value)
        return True
    except ValueError:
        return False

# Filter out non-numeric values in the 'Poverty' column
combined_data_poverty = combined_data_poverty[combined_data_poverty['Poverty'].apply(is_numeric)]

# Convert 'Poverty' column to numeric
combined_data_poverty['Poverty'] = pd.to_numeric(combined_data_poverty['Poverty'])

# Central tendency measures
mean_poverty = combined_data_poverty.groupby('Year')['Poverty'].mean()
median_poverty = combined_data_poverty.groupby('Year')['Poverty'].median()

# Summary statistics
summary_stats = combined_data_poverty.groupby('Year')['Poverty'].describe()

# Print results
print("Mean Poverty:\n", mean_poverty)
print("\nMedian Poverty:\n", median_poverty)
print("\nSummary Statistics:\n", summary_stats)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Extract relevant columns for Total Enrollment
relevant_cols_enrollment = ['# Asian', '# Hispanic', 'TotalEnrollment']

# Concatenate the datasets
combined_data_enrollment = pd.concat([df1[relevant_cols_enrollment], df2[relevant_cols_enrollment], df3[relevant_cols_enrollment]], keys=['df1', 'df2', 'df3'])

# Convert columns to numeric
combined_data_enrollment['# Asian'] = pd.to_numeric(combined_data_enrollment['# Asian'])
combined_data_enrollment['# Hispanic'] = pd.to_numeric(combined_data_enrollment['# Hispanic'])
combined_data_enrollment['TotalEnrollment'] = pd.to_numeric(combined_data_enrollment['TotalEnrollment'])

# Plot pie charts for each dataset
datasets = combined_data_enrollment.index.get_level_values(0).unique()

plt.figure(figsize=(15, 5))

for i, dataset in enumerate(datasets, 1):
    plt.subplot(1, 3, i)
    data_subset = combined_data_enrollment.loc[dataset]
    total_enrollment = data_subset['TotalEnrollment'].sum()
    other_enrollment = total_enrollment - data_subset[['# Asian', '# Hispanic']].sum().sum()
    pie_data = [data_subset['# Asian'].sum(), data_subset['# Hispanic'].sum(), other_enrollment]
    labels = ['# Asian', '# Hispanic', 'Other']
    plt.pie(pie_data, labels=labels, autopct=lambda p: f'{p:.1f}%\n({int(p * total_enrollment / 100)})')
    plt.title(f'Dataset {dataset} - Total Enrollment: {total_enrollment}')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Extract relevant columns for Total Enrollment
relevant_cols_enrollment = ['# Asian', '# Hispanic', 'TotalEnrollment']

# Concatenate the datasets
combined_data_enrollment = pd.concat([df1[relevant_cols_enrollment], df2[relevant_cols_enrollment], df3[relevant_cols_enrollment]], keys=['df1', 'df2', 'df3'])

# Convert columns to numeric
combined_data_enrollment['# Asian'] = pd.to_numeric(combined_data_enrollment['# Asian'])
combined_data_enrollment['# Hispanic'] = pd.to_numeric(combined_data_enrollment['# Hispanic'])
combined_data_enrollment['TotalEnrollment'] = pd.to_numeric(combined_data_enrollment['TotalEnrollment'])

# Calculate correlation coefficients
correlation_matrix = combined_data_enrollment.corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

correlation_asian_enrollment = combined_data_enrollment['# Asian'].corr(combined_data_enrollment['TotalEnrollment'])
print(f"\nCorrelation between '# Asian' and 'TotalEnrollment': {correlation_asian_enrollment}")

# Similarly, correlation between '# Hispanic' and 'TotalEnrollment'
correlation_hispanic_enrollment = combined_data_enrollment['# Hispanic'].corr(combined_data_enrollment['TotalEnrollment'])
print(f"Correlation between '# Hispanic' and 'TotalEnrollment': {correlation_hispanic_enrollment}")
```

This also suggests a strong positive correlation between the number of Hispanic students and the total enrollment. As the number of Hispanic students increases, the total enrollment tends to increase.
The positive correlations imply that schools with higher numbers of Asian or Hispanic students tend to have larger total enrollments.
